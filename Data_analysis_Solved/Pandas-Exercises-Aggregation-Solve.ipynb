# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
df=pd.read_csv('Occupation.csv',sep='|')
df
```

### Step 3. Assign it to a variable called users.

```python
users=df
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
df['gender_encoded'] = np.where(df['gender'] == 'M', 1, 0)
df
```

```python
occupation_stats =df.groupby('occupation').agg(
    total_people=('gender', 'count'), 
    male_people=('gender_encoded', 'sum')
)
```

```python
occupation_stats.head()
```

```python
occupation_stats['male_ratio']=occupation_stats['male_people']/occupation_stats['total_people']
```

```python
occupation_stats.head(2)
```

```python
occupation_stats.sort_values(by='male_ratio', ascending=False)
```

```python
df.head(2)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(['occupation'])['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender'])['age'].mean().reset_index()
```

### Step 8.  For each occupation present the percentage of women and men

```python
occupation_stats['female_people'] = occupation_stats['total_people'] - occupation_stats['male_people']
occupation_stats
```

```python
occupation_stats['male_percentage'] = (occupation_stats['male_people'] / occupation_stats['total_people']) * 100
occupation_stats['female_percentage'] = (occupation_stats['female_people'] / occupation_stats['total_people']) * 100
```

```python
occupation_stats[['male_percentage', 'female_percentage']]
```
